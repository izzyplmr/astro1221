```python
def mag_to_flux(magnitude, zero_point=25.0):
    flux = 10 ** ((zero_point - magnitude) / 2.5)
    return flux
print(mag_to_flux(5.0))
print(mag_to_flux(5.0, 26.0))
print(mag_to_flux(5.0) / mag_to_flux(26.0))
```

```python
def function_with_print(x=2):
    """This function only prints."""
    result = x * 2
    print(result)  # Shows on screen but doesn't send back

def function_with_return(x=2):
    """This function returns a value."""
    result = x * 2
    return result  # Sends the value back
```
